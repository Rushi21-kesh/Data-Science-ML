# Data Cleaning
In data cleaning for this dataframe we Drop serial No. columns and in column Chance of Admit                          
if value is < 0.5 make it '0'                     
           else make it '1' .

```python
#drop Serial No. column
df=df.drop('Serial No.',axis=1)
```

```python
df.head()
```

```python
#if value of Chance of Admit is < 0.5 make it '0' else make it '1'
df['Chance of Admit ']=np.where(df['Chance of Admit '] >= 0.5,'1','0')
```

```python
df.head()
```

# Building a Naive Bayes model


## Train Test Split

```python
X=df.iloc[:,:7]
X.head()
```

```python
Y=df['Chance of Admit ']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_text=train_test_split(X,Y,test_size=0.20,random_state=1)
```

## Training and Predicting

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
acc=model.score(x_test,y_text)
```

```python
print("Accuracy is {0:.2f} %".format(acc*100))
```
