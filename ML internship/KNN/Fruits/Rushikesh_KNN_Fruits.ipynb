### Building a Machine Learning model 

```python
#import model
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
```

```python
# spliting data for Training and Testing where 'x' training data and 'y' is testing data
X=df.iloc[:,3:]
Y=df['fruit_name']
```

```python
X.head()
```

```python
Y.head()
```

```python
#split data into training and testing by using train_test_split() method
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=20,random_state=0)
Scalar=MinMaxScaler()
X_train=Scalar.fit_transform(X_train)
X_test=Scalar.transform(X_test)
```

```python
#check data is split right or not
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#train model to predict type of fruit 
model = KNeighborsClassifier(n_neighbors=13)
model.fit(X_train,Y_train)
print("MODEL TRAINED SUCESSFULLY")
```

```python
(model.predict([[192,8.4,7.3,0.55]]))[0]
```

```python
#here we find the accuracy
accuracy=[]
for K in range(1,35,2):
    model = KNeighborsClassifier(n_neighbors=K,metric='minkowski',leaf_size=30,p=2,weights='uniform')
    model.fit(X_train,Y_train)
    Y_predict=model.predict(X_test)
    acc=metrics.accuracy_score(Y_test,Y_predict)
    accuracy.append(acc)
print(accuracy)
```

```python
#from these plot we can say that 3 is a best K-value
K=list(range(1,35,2))
plt.figure(figsize=(10,6))
plt.plot(K,accuracy,'*-')
plt.xlabel("Value of K")
plt.ylabel("Accuracy Score")
```

```python
ac=max(accuracy)
print("Accuracy is : {:.3f}".format(ac*100),"%" )
```
